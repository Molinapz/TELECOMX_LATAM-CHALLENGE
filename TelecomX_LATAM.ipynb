# Telecom X LATAM - Analisis de churn

Este notebook desarrolla el challenge completo de Telecom X con un flujo de **Extraccion, Transformacion, Carga y Analisis**. El objetivo es entender que factores estan mas relacionados con la evasion de clientes y dejar una base lista para futuras etapas de ciencia de datos.

## Objetivos del notebook
- Cargar y normalizar los datos JSON.
- Limpiar inconsistencias y preparar variables utiles.
- Realizar un analisis exploratorio con visualizaciones.
- Redactar conclusiones y recomendaciones accionables.


```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from pathlib import Path
from IPython.display import display

sns.set_theme(style="whitegrid", palette="Blues")
plt.rcParams["figure.figsize"] = (12, 6)
plt.rcParams["axes.titlesize"] = 14
plt.rcParams["axes.labelsize"] = 11
```

## 1. Extraccion

La fuente original del challenge esta en formato JSON. Primero cargamos los datos desde la URL raw del repositorio base; si no esta disponible, el notebook usa una copia local llamada `TelecomX_Data.json`.


```python
DATA_URL = "https://raw.githubusercontent.com/ingridcristh/challenge2-data-science-LATAM/main/TelecomX_Data.json"
LOCAL_DATA = Path("TelecomX_Data.json")

try:
    telecom_raw = pd.read_json(DATA_URL)
    data_source = DATA_URL
except Exception as exc:
    if not LOCAL_DATA.exists():
        raise FileNotFoundError(
            "No se pudo cargar la API y tampoco existe un archivo local llamado TelecomX_Data.json"
        ) from exc
    telecom_raw = pd.read_json(LOCAL_DATA)
    data_source = str(LOCAL_DATA.resolve())

print(f"Fuente utilizada: {data_source}")
print(f"Registros cargados: {telecom_raw.shape[0]}")
telecom_raw.head(3)
```

```python
telecom_df = pd.json_normalize(telecom_raw.to_dict(orient="records"))
telecom_df.columns = (
    telecom_df.columns.str.replace("customer.", "", regex=False)
    .str.replace("phone.", "", regex=False)
    .str.replace("internet.", "", regex=False)
    .str.replace("account.", "", regex=False)
    .str.replace("Charges.", "Charges_", regex=False)
)

rename_map = {
    "customerID": "customer_id",
    "Churn": "churn",
    "gender": "gender",
    "SeniorCitizen": "senior_citizen",
    "Partner": "partner",
    "Dependents": "dependents",
    "tenure": "tenure_months",
    "PhoneService": "phone_service",
    "MultipleLines": "multiple_lines",
    "InternetService": "internet_service",
    "OnlineSecurity": "online_security",
    "OnlineBackup": "online_backup",
    "DeviceProtection": "device_protection",
    "TechSupport": "tech_support",
    "StreamingTV": "streaming_tv",
    "StreamingMovies": "streaming_movies",
    "Contract": "contract",
    "PaperlessBilling": "paperless_billing",
    "PaymentMethod": "payment_method",
    "Charges_Monthly": "monthly_charges",
    "Charges_Total": "total_charges",
}

telecom_df = telecom_df.rename(columns=rename_map)
display(telecom_df.head())
print(f"Filas: {telecom_df.shape[0]} | Columnas: {telecom_df.shape[1]}")
```

## 2. Transformacion

En esta etapa revisamos la estructura del dataset, detectamos incoherencias y preparamos una version limpia para el analisis.


```python
telecom_df.info()

null_report = telecom_df.isna().sum().rename("null_values").to_frame()
blank_report = (
    telecom_df.select_dtypes(include="object")
    .apply(lambda col: col.astype(str).str.strip().eq("").sum())
    .rename("blank_strings")
    .to_frame()
)
quality_report = null_report.join(blank_report, how="left").fillna(0).astype(int)

display(quality_report[quality_report.sum(axis=1) > 0])
print("Duplicados por customer_id:", telecom_df.duplicated(subset="customer_id").sum())
```

```python
df = telecom_df.copy()

object_columns = df.select_dtypes(include="object").columns
df[object_columns] = df[object_columns].apply(lambda col: col.str.strip())

df["churn"] = df["churn"].replace("", np.nan)
df["total_charges"] = pd.to_numeric(df["total_charges"], errors="coerce")

mask_zero_tenure = df["total_charges"].isna() & df["tenure_months"].eq(0)
df.loc[mask_zero_tenure, "total_charges"] = 0

df = df.dropna(subset=["churn"]).copy()
df["senior_citizen"] = df["senior_citizen"].astype(int)
df["churn_flag"] = df["churn"].map({"Yes": 1, "No": 0})
df["daily_charges"] = df["monthly_charges"] / 30

service_columns = [
    "phone_service",
    "multiple_lines",
    "online_security",
    "online_backup",
    "device_protection",
    "tech_support",
    "streaming_tv",
    "streaming_movies",
]

df["total_services"] = (
    df[service_columns].eq("Yes").sum(axis=1)
    + df["internet_service"].ne("No").astype(int)
)
df["senior_citizen_label"] = df["senior_citizen"].map({0: "No", 1: "Yes"})

print("Registros originales:", telecom_df.shape[0])
print("Registros sin churn removidos:", telecom_df["churn"].astype(str).str.strip().eq("").sum())
print("Total charges vacios corregidos con tenure = 0:", int(mask_zero_tenure.sum()))
print("Dataset limpio:", df.shape)
display(df.head())
```

### Resumen de limpieza

- Se cargaron **7,267** clientes desde el JSON original.
- Se detectaron **224** registros sin valor en `churn`, por lo que se excluyeron del analisis descriptivo.
- Se encontraron **11** registros con `total_charges` vacio y `tenure_months = 0`; se imputaron con `0`.
- No se encontraron duplicados por `customer_id`.
- El dataset final quedo con **7,043 clientes** y variables adicionales como `daily_charges`, `churn_flag` y `total_services`.


```python
numeric_columns = [
    "tenure_months",
    "monthly_charges",
    "total_charges",
    "daily_charges",
    "total_services",
]

display(df[numeric_columns].describe().T.round(2))

churn_summary = (
    df.groupby("churn")[numeric_columns]
    .agg(["mean", "median", "std"])
    .round(2)
)
display(churn_summary)
```

Los clientes que cancelan muestran un patron claro: menor antiguedad promedio (**17.98** vs **37.57** meses) y mayor cargo mensual promedio (**74.44** vs **61.27**). Aunque el gasto total promedio es menor entre quienes cancelan (**1531.80** vs **2549.91**), esto se explica por el menor tiempo de permanencia.


## 3. Carga y analisis

Con los datos limpios, pasamos al analisis exploratorio para encontrar perfiles y comportamientos con mayor riesgo de churn.


```python
churn_distribution = (
    df["churn"]
    .value_counts()
    .rename_axis("churn")
    .reset_index(name="clients")
)
churn_distribution["share"] = (churn_distribution["clients"] / len(df) * 100).round(2)
display(churn_distribution)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
sns.barplot(data=churn_distribution, x="churn", y="clients", palette=["#4C78A8", "#F58518"], ax=axes[0])
axes[0].set_title("Distribucion de churn")
axes[0].set_xlabel("")
axes[0].set_ylabel("Clientes")

axes[1].pie(
    churn_distribution["clients"],
    labels=churn_distribution["churn"],
    autopct="%1.1f%%",
    startangle=90,
    colors=["#4C78A8", "#F58518"],
)
axes[1].set_title("Proporcion de churn")

plt.tight_layout()
plt.show()
```

La tasa global de churn fue **26.54%**, por lo que aproximadamente 1 de cada 4 clientes del dataset termino cancelando el servicio.


```python
def churn_rate_table(data, column):
    summary = (
        data.groupby(column)
        .agg(clientes=("churn_flag", "size"), churn_rate=("churn_flag", "mean"))
        .sort_values("churn_rate", ascending=False)
        .reset_index()
    )
    summary["churn_rate"] = (summary["churn_rate"] * 100).round(2)
    return summary

categorical_features = {
    "contract": "Tipo de contrato",
    "payment_method": "Metodo de pago",
    "internet_service": "Servicio de internet",
    "senior_citizen_label": "Adulto mayor",
}

fig, axes = plt.subplots(2, 2, figsize=(16, 10))
for ax, (feature, title) in zip(axes.flat, categorical_features.items()):
    summary = churn_rate_table(df, feature)
    sns.barplot(data=summary, x=feature, y="churn_rate", palette="Blues_r", ax=ax)
    ax.set_title(title)
    ax.set_xlabel("")
    ax.set_ylabel("Churn (%)")
    ax.tick_params(axis="x", rotation=20)
    for idx, row in summary.reset_index(drop=True).iterrows():
        ax.text(idx, row["churn_rate"] + 0.8, f"{row['churn_rate']:.1f}%", ha="center", fontsize=9)

plt.tight_layout()
plt.show()

display(churn_rate_table(df, "contract"))
display(churn_rate_table(df, "payment_method"))
display(churn_rate_table(df, "internet_service"))
```

### Insights por variables categoricas

- Los contratos **Month-to-month** concentran el mayor riesgo de churn (**41.32%**), muy por encima de **One year** (**10.93%**) y **Two year** (**2.75%**).
- El metodo de pago con mayor evasion es **Electronic check** (**43.80%**).
- Los clientes con **Fiber optic** presentan un churn de **40.56%**, contra **18.45%** en **DSL**.
- El genero no muestra diferencias importantes, por lo que no parece ser una variable prioritaria para explicar la fuga.
- Los clientes adultos mayores tienen una tasa de churn de **40.27%**, frente a **22.89%** en el resto.


```python
service_features = {
    "online_security": "Seguridad en linea",
    "tech_support": "Soporte tecnico",
    "paperless_billing": "Factura digital",
    "dependents": "Dependientes",
}

fig, axes = plt.subplots(2, 2, figsize=(16, 10))
for ax, (feature, title) in zip(axes.flat, service_features.items()):
    summary = churn_rate_table(df, feature)
    sns.barplot(data=summary, x=feature, y="churn_rate", palette="crest", ax=ax)
    ax.set_title(title)
    ax.set_xlabel("")
    ax.set_ylabel("Churn (%)")
    ax.tick_params(axis="x", rotation=20)
    for idx, row in summary.reset_index(drop=True).iterrows():
        ax.text(idx, row["churn_rate"] + 0.8, f"{row['churn_rate']:.1f}%", ha="center", fontsize=9)

plt.tight_layout()
plt.show()
```

Las variables de retencion muestran una senal fuerte: los clientes sin **online_security** (**41.77%**) y sin **tech_support** (**41.64%**) cancelan mucho mas que quienes si cuentan con estos servicios. Ademas, los clientes sin dependientes (**31.28%**) y quienes usan factura digital (**33.57%**) presentan un riesgo mayor.


```python
fig, axes = plt.subplots(2, 2, figsize=(16, 10))

sns.boxplot(data=df, x="churn", y="tenure_months", palette=["#4C78A8", "#F58518"], ax=axes[0, 0])
axes[0, 0].set_title("Antiguedad por churn")

sns.boxplot(data=df, x="churn", y="monthly_charges", palette=["#4C78A8", "#F58518"], ax=axes[0, 1])
axes[0, 1].set_title("Cargo mensual por churn")

sns.histplot(data=df, x="tenure_months", hue="churn", bins=30, stat="density", common_norm=False, element="step", ax=axes[1, 0])
axes[1, 0].set_title("Distribucion de antiguedad")

sns.histplot(data=df, x="monthly_charges", hue="churn", bins=30, stat="density", common_norm=False, element="step", ax=axes[1, 1])
axes[1, 1].set_title("Distribucion de cargos mensuales")

plt.tight_layout()
plt.show()

tenure_bins = pd.cut(
    df["tenure_months"],
    bins=[-1, 12, 24, 48, 72],
    labels=["0-12", "13-24", "25-48", "49-72"],
)
monthly_bins = pd.cut(
    df["monthly_charges"],
    bins=[18, 35, 70, 90, 120],
    labels=["18-35", "35-70", "70-90", "90-120"],
    include_lowest=True,
    right=False,
)

tenure_risk = (
    df.assign(tenure_band=tenure_bins)
    .groupby("tenure_band")
    .agg(clientes=("churn_flag", "size"), churn_rate=("churn_flag", "mean"))
    .assign(churn_rate=lambda x: (x["churn_rate"] * 100).round(2))
)
monthly_risk = (
    df.assign(monthly_band=monthly_bins)
    .groupby("monthly_band")
    .agg(clientes=("churn_flag", "size"), churn_rate=("churn_flag", "mean"))
    .assign(churn_rate=lambda x: (x["churn_rate"] * 100).round(2))
)

display(tenure_risk)
display(monthly_risk)
```

### Insights por variables numericas

- El segmento con **0 a 12 meses** de antiguedad alcanza un churn de **47.44%**.
- Entre **49 y 72 meses**, el churn baja a **9.51%**.
- Los cargos mensuales entre **70 y 90** concentran el mayor churn (**37.95%**), seguidos por el rango **90 a 120** (**32.86%**).
- Los clientes con cargos mensuales bajos (**18 a 35**) presentan un churn mucho menor (**10.86%**).


```python
corr_df = df[
    [
        "churn_flag",
        "senior_citizen",
        "tenure_months",
        "monthly_charges",
        "total_charges",
        "daily_charges",
        "total_services",
    ]
].copy()

binary_features = [
    "partner",
    "dependents",
    "paperless_billing",
    "phone_service",
    "multiple_lines",
    "online_security",
    "online_backup",
    "device_protection",
    "tech_support",
    "streaming_tv",
    "streaming_movies",
]

for col in binary_features:
    corr_df[col] = df[col].map({"Yes": 1, "No": 0, "No internet service": 0, "No phone service": 0})

plt.figure(figsize=(14, 10))
sns.heatmap(corr_df.corr(numeric_only=True), cmap="Blues", center=0, annot=False)
plt.title("Matriz de correlacion")
plt.show()
```

## 4. Informe final

### Introduccion
Telecom X enfrenta una fuga importante de clientes y necesita entender que perfiles y comportamientos estan mas asociados a la cancelacion del servicio. El objetivo de este analisis fue construir una base limpia, explorar patrones y generar insights utiles para acciones de retencion.

### Limpieza y tratamiento de datos
- Se importaron los datos en formato JSON y se aplanaron las estructuras anidadas para convertirlas en un DataFrame tabular.
- Se detectaron 224 registros sin valor en `churn` y se excluyeron del analisis porque no permitian etiquetar la evasion.
- Se corrigieron 11 registros con `total_charges` vacio y `tenure_months = 0`, asignando `0` como valor coherente.
- Se crearon variables derivadas como `daily_charges`, `churn_flag` y `total_services` para ampliar el analisis.

### Analisis exploratorio de datos
- La tasa global de churn fue **26.54%**.
- El perfil con mayor riesgo aparece en clientes con contratos **Month-to-month**, pagos por **Electronic check** y servicio de internet **Fiber optic**.
- Las variables de permanencia y precio muestran una senal clara: quienes cancelan suelen tener menor antiguedad y cargos mensuales mas altos.
- Los servicios de valor agregado asociados a soporte y seguridad parecen funcionar como factores de retencion.

### Conclusiones e insights
1. La fuga se concentra en clientes nuevos, de baja permanencia y con contratos flexibles.
2. El precio mensual alto, especialmente en clientes con fibra optica, parece aumentar la probabilidad de cancelacion.
3. Los servicios complementarios de seguridad y soporte tecnico se relacionan con mayor permanencia.
4. Variables como genero o servicio telefonico muestran diferencias pequenas y no deberian ser prioridad en una estrategia comercial.

### Recomendaciones
1. Crear campanas de retencion para clientes con menos de 12 meses y contrato `Month-to-month`, con incentivos para migrar a planes anuales.
2. Revisar la propuesta de valor del segmento **Fiber optic**, especialmente en combinacion con cargos mensuales altos.
3. Disenar bundles con **online_security** y **tech_support** para clientes con mayor riesgo de fuga.
4. Auditar la experiencia de pago en **Electronic check** y el flujo de factura digital para detectar fricciones.
5. Priorizar acciones preventivas en clientes sin pareja o sin dependientes, ya que muestran una tasa de churn superior al promedio.
